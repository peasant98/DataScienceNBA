**Westbrook Exposed**:

This bit of Python code here takes from a large data of NBA players, both present and past, to determine the usage percentage for 20+ PPG scorers over the years. The first graph shown below is intended to be misleading; to show that Russell Westbrook from 2017 was just your average high scorer from the NBA.

*However*, accurately potraying the data reveals the reals facts about him as a player that year......

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
# Matthew Strong 
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/peasant98/TheNBACSV/master/nbaNew.csv')
real = df.dropna(subset=['PlayerName', 'PTS', 'G', 'SeasonStart'])
pd.options.mode.chained_assignment = None 

real['PPG'] = round(real['PTS']/real['G'],2)

pointPlayers = real[[ 'PlayerName',  'TOV%', 'SeasonStart', 'PER', 'PPG', '3PA', 'USG%', 'FG%']]
#pointPlayers[pointPlayers.index.duplicated()]
aYear = pointPlayers.loc[(pointPlayers['SeasonStart']==2017)&(pointPlayers['PPG']>=20)]
#aYear = pointPlayers[(pointPlayers.SeasonStart==2017)&(pointPlayers.PPG>=20)]



bYear = pointPlayers.loc[(pointPlayers['SeasonStart']<=2016)&(pointPlayers['PPG']>=20) &(pointPlayers['SeasonStart']>=1980)]

cYear = bYear.sort_values(by='SeasonStart')

z = aYear[['USG%']].values
y = cYear[['USG%']].values


plt.plot(z, label='2017')


plt.plot(y, label='1980-2016')
plt.legend(loc='lower right')
plt.title("Usage % of 20+ PPG Scorers from 1980-2017.")
plt.ylabel("Usage Percent")
plt.ylim(30,38)
plt.xlabel("No. of Players")
plt.show()


```

Doesn't look like anything, right? Random data, random, useless.

```python
plt.plot(z, label='2017')
plt.plot(y, label='1980-2016')

plt.title("Usage % of 20+ PPG Scorers from 1980-2017.")
plt.ylabel("Usage Percent")
o = [41.7]*900
plt.plot(o, label=' 2017 Russell Westbrook (41.7%)')
plt.legend(loc='lower right')
plt.xlabel("No. of Players")
plt.show()
```

Hopefully, this graph speaks for itself. Some players really enjoyed using the ball to score high amounts, but Westbrook **really** used the basketball that year.
